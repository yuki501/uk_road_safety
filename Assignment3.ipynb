## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
df_a = spark.read.format("csv") \
  .option("inferSchema", "true") \
  .option("header", "true") \
  .option("sep", ",") \
  .load("/FileStore/tables/Accident_Information.csv")

df_v = spark.read.format("csv") \
  .option("inferSchema", "true") \
  .option("header", "true") \
  .option("sep", ",") \
  .load("/FileStore/tables/Vehicle_Information.csv")

display(df_a)
```

```
display(df_v)
```

```
# Create tables

df_a.createOrReplaceTempView("accident_information")
df_v.createOrReplaceTempView("vehicle_information")
```

```
%sql

select a.Accident_Severity, count(*)
from accident_information a
group by a.Accident_Severity
```

```
%sql

SELECT Light_conditions, AVG(Speed_limit) as AVG_Road_Speed_Limit
FROM accident_information
WHERE NOT Light_conditions = 'Data missing or out of range'
GROUP BY Light_conditions
ORDER BY AVG_Road_Speed_Limit;
```

```
%sql

SELECT Light_conditions, Urban_or_Rural_Area, count(*)
FROM accident_information
WHERE NOT Light_conditions = 'Data missing or out of range' AND NOT Urban_or_Rural_Area = 'Unallocated'
GROUP BY Light_conditions, Urban_or_Rural_Area
ORDER BY Light_conditions
```

```
%sql

SELECT a.Accident_Severity, v.Age_Band_of_Driver, COUNT(*) as Number_of_Accidents
FROM accident_information a
INNER JOIN vehicle_information v ON v.Accident_Index=a.Accident_Index
WHERE NOT v.Age_Band_of_Driver='Data missing or out of range'
GROUP BY a.Accident_Severity, v.Age_Band_of_Driver
ORDER BY Number_of_Accidents DESC;
```

```
%sql

SELECT x.make, count(*) as make_count, rank() over (order by count(*) desc) as rank, sum(Case When x.Accident_Severity = 'Serious' OR x.Accident_Severity = 'Fatal' Then 1 Else 0 End)/count(*)*1.0 as serious_percent
FROM
  (SELECT a.Accident_Severity, v.make
  FROM vehicle_information v
  INNER JOIN accident_information a ON a.Accident_Index=v.Accident_Index
  WHERE NOT v.make='NULL') x
GROUP BY x.make limit 50
```

```
df_heatmap = sqlContext.sql("SELECT CAST(Longitude as float), CAST(Latitude as float), count(*) as accident_rate FROM accident_information WHERE NOT Longitude = 'NA' AND NOT Latitude = 'NA' GROUP BY Longitude, Latitude")
df_hm = df_heatmap.toPandas()
```

```
import matplotlib.pyplot as plt
import pandas as pd
x = df_hm['Longitude'].tolist()
y = df_hm['Latitude'].tolist()
c = df_hm['accident_rate'].tolist()

fig, ax = plt.subplots()
fig.set_size_inches(5,10)
ax.scatter(x=x,y=y,s=0.001, c=c)
display(fig)
```

```
targetDf = sqlContext.sql("SELECT a.Accident_Severity, a.Number_of_Casualties, a.Number_of_Vehicles, a.Road_Surface_Conditions, a.Speed_limit, v.Age_Band_of_Driver, v.Age_of_Vehicle, v.Vehicle_Manoeuvre, v.Vehicle_Type, v.Sex_of_Driver FROM accident_information a INNER JOIN vehicle_information v ON v.Accident_Index=a.Accident_Index WHERE NOT a.Road_Surface_Conditions='Data missing or out of range' AND NOT a.Speed_Limit='NA' AND NOT v.Age_Band_of_Driver='Data missing or out of range' AND NOT v.Age_of_Vehicle = 'NA' AND NOT v.Sex_of_Driver = 'Not known' AND NOT v.Sex_of_Driver = 'Data missing or out of range' AND NOT v.Vehicle_Type = 'Data missing or out of range' AND NOT v.Vehicle_Manoeuvre = 'Data missing or out of range' ")
display(targetDf)
```

```
from pyspark.sql.functions import when

df = targetDf.withColumn("Severity", \
              when(targetDf["Accident_Severity"] == 'Fatal', 'Serious').otherwise(targetDf["Accident_Severity"]))
display(df)


```

```
df_2 = df.select(df.Severity, df.Number_of_Casualties, df.Number_of_Vehicles, df.Road_Surface_Conditions, df.Speed_limit.cast("INT"), df.Age_Band_of_Driver, df.Vehicle_Manoeuvre, df.Vehicle_Type, df.Age_of_Vehicle.cast("INT"), df.Sex_of_Driver)
df_2.schema
```

```
from pyspark.ml import Pipeline
from pyspark.ml.feature import OneHotEncoderEstimator, StringIndexer, VectorAssembler
categoricalColumns =['Road_Surface_Conditions', 'Age_Band_of_Driver', 'Vehicle_Manoeuvre', 'Vehicle_Type', 'Sex_of_Driver']
stages = [] # stages in our Pipeline
for categoricalCol in categoricalColumns:
    # Category Indexing with StringIndexer
    stringIndexer = StringIndexer(inputCol=categoricalCol, outputCol=categoricalCol + "Index")
    # Use OneHotEncoder to convert categorical variables into binary SparseVectors
    # encoder = OneHotEncoderEstimator(inputCol=categoricalCol + "Index", outputCol=categoricalCol + "classVec")
    encoder = OneHotEncoderEstimator(inputCols=[stringIndexer.getOutputCol()], outputCols=[categoricalCol + "classVec"])
    # Add stages.  These are not run here, but will run all at once later on.
    stages += [stringIndexer, encoder]
```

```
# Convert label into label indices using the StringIndexer
label_stringIdx = StringIndexer(inputCol="Severity", outputCol="label")
stages += [label_stringIdx]
```

```
numericCols = ["Number_of_Casualties", "Number_of_Vehicles", "Speed_limit", "Age_of_Vehicle"]
assemblerInputs = [c + "classVec" for c in categoricalColumns] + numericCols
assembler = VectorAssembler(inputCols=assemblerInputs, outputCol="features")
stages += [assembler]
```

```
from pyspark.ml.classification import LogisticRegression
  
partialPipeline = Pipeline().setStages(stages)
pipelineModel = partialPipeline.fit(df_2)
preppedDataDF = pipelineModel.transform(df_2)
```

```
# Fit model to prepped data
lrModel = LogisticRegression().fit(preppedDataDF)

# ROC for training data
display(lrModel, preppedDataDF, "ROC")
```

```
### Randomly split data into training and test sets. set seed for reproducibility
(trainingData, testData) = preppedDataDF.randomSplit([0.8, 0.2], seed=100)
trainingData.createOrReplaceTempView("trainingData")
```

```
%sql

SELECT label, count(*)
FROM trainingData
GROUP BY label
```

```
trainingData.createOrReplaceTempView("trainingData")

df_slight = sqlContext.sql("SELECT * FROM trainingData WHERE label=0 ORDER BY RAND() LIMIT 184812")
df_serious = sqlContext.sql("SELECT * FROM trainingData WHERE label=1")

df_train = df_slight.unionAll(df_serious)
```

```
# Create initial LogisticRegression model
lr = LogisticRegression(labelCol="label", featuresCol="features", maxIter=10)

# Train model with Training Data
lrModel = lr.fit(df_train)
```

```
predictions = lrModel.transform(testData)
selected = predictions.select("label", "prediction", "probability")
display(selected)
```

```
display(lrModel, testData, "ROC")
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Evaluate model
evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction")
evaluator.evaluate(predictions)
```

```
selected.createOrReplaceTempView("selected")
```

```
%sql

select sum(Case When label = prediction Then 1 Else 0 End)/count(*)*1.0 as accuracy_serious  from (select * from selected where label=1 )
```

```
%sql

select sum(Case When label = prediction Then 1 Else 0 End)/count(*)*1.0 as accuracy_slight from (select * from selected where label=0 )
```
